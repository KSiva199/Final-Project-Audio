```python
import librosa as librosa
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load audio file
y, sr = librosa.load('./american_m_Josh_20230405130012.wav')

librosa.display.waveshow(y,sr=sr,color='black')
```

```python
# Compute spectrogram of original signal
S = librosa.stft(y)


# Remove leading/trailing silence
y_trimmed, idx = librosa.effects.trim(y)
# Compute spectrogram of trimmed signal
S_trimmed = librosa.stft(y_trimmed)


# segments an audio signal into regions of active sound and silence
# librosa.effects.split()
#Remove out of tunes
#librosa.effects.estimate_tuning()


# Plot original and trimmed spectrograms
plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
librosa.display.specshow(librosa.amplitude_to_db(abs(S), ref=np.max), sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Original signal')
plt.subplot(2, 1, 2)
librosa.display.specshow(librosa.amplitude_to_db(abs(S_trimmed), ref=np.max), sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Trimmed signal')
plt.tight_layout()
plt.show()
```
