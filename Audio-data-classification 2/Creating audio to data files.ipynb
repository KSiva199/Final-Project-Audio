```python
import librosa as librosa
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
import os
import sys
from pathlib import Path
```

```python
for digit in range(10):
    numpy_dir = f'Audio np arrays/{digit}/'
    png_dir = f'Spectro_imgs/{digit}/'

    # Directory of wav files
    dir_list = os.listdir(f"./audio/{digit}/")
    
    for i,audio_file in enumerate(dir_list):
         #create mel spectrogram
        samples, sample_rate = librosa.load(f"./audio/{digit}/{audio_file}", sr=None)
        # # Set the desired number of mel frequency bands and FFT window size
        # n_mels = 128
        # n_fft = 2048

        melsgram = librosa.feature.melspectrogram(y=samples,sr=sample_rate)
        # melsgram = librosa.feature.melspectrogram(y=samples, sr=sample_rate, n_mels=n_mels, n_fft=n_fft)
        # Resize the mel spectrogram to a fixed size
        fixed_size = librosa.util.fix_length(melsgram, size=100, axis=1)


        # path to save numpy arr of spectro
        if not os.path.exists(numpy_dir):
            Path(numpy_dir).mkdir(exist_ok=True,parents=True)
        numpy_destination=os.path.join(numpy_dir,f"{audio_file[:-3]}")

        # path to save spectro images
        if not os.path.exists(png_dir):
            Path(png_dir).mkdir(exist_ok=True,parents=True)
        png_destination=os.path.join(png_dir,f"{audio_file[:-3]}png")

        np.save(numpy_destination,fixed_size)
        matplotlib.image.imsave(png_destination, fixed_size)
        print( f'{digit}: {fixed_size.shape} saved')

```
